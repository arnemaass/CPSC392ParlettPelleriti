# Doing Good Data Visualization

* What am I trying to *communicate* ?
    * Which elements of my graph support that?
    * Which elements of my graph detract from that?
* Who is my *audience* ?
    * What context will people see this?
* Is my graph *accessible* ?
    * colorblindness
    * visual impairment
    * cultural context
